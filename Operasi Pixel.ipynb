# Import Library

```python
import cv2
import matplotlib.pyplot as plt
```

# Membaca Gambar

```python
img = cv2.imread('backlight.jpg')
```

# Convert BGR to RGB

```python
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

# Convert RGB to Grayscale

```python
gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)
```

# Gray Dipercerah

```python
gray_bright = cv2.convertScaleAbs(gray, alpha=1, beta=80)
```

# Gray Diperkontras

```python
gray_contrast = cv2.convertScaleAbs(gray, alpha=2, beta=0)
```

# Gray Dipercerah dan Diperkontras

```python
gray_bright_contrast = cv2.convertScaleAbs(gray, alpha=2, beta=80)
```

# Menampilkan Hasil

```python
fig, axs = plt.subplots(2,3, figsize=(15,10))

# 1. Gambar Asli
axs[0,0].imshow(rgb)
axs[0,0].set_title("Gambar Asli")
axs[0,0].axis('off')

# 2. Gambar Gray
axs[0,1].imshow(gray, cmap='gray')
axs[0,1].set_title("Gambar Gray")
axs[0,1].axis('off')

# 3. Gray Dipercerah
axs[0,2].imshow(gray_bright, cmap='gray')
axs[0,2].set_title("Gray Dipercerah")
axs[0,2].axis('off')

# 4. Gray Diperkontras
axs[1,0].imshow(gray_contrast, cmap='gray')
axs[1,0].set_title("Gray Diperkontras")
axs[1,0].axis('off')

# 5. Gray Dipercerah dan Diperkontras
axs[1,1].imshow(gray_bright_contrast, cmap='gray')
axs[1,1].set_title("Gray Dipercerah & Diperkontras")
axs[1,1].axis('off')

# menghapus subplot kosong
axs[1,2].axis('off')

plt.tight_layout()
plt.show()
```
